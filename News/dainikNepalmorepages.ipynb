```python
import requests
from bs4 import BeautifulSoup
from datetime import date
import pandas as pd
import re
import os
```

```python
url =['https://www.dainiknepal.com/section/latest-news',
'https://www.dainiknepal.com/section/views',
'https://www.dainiknepal.com/section/kala',
'https://www.dainiknepal.com/section/diaspora',
'https://www.dainiknepal.com/section/market',
'https://www.dainiknepal.com/section/sports',
'https://www.dainiknepal.com/section/rochak',
'https://www.dainiknepal.com/section/health',]

print(url)
```

```python
for i in url:
    response= requests.get(i)
    soup = BeautifulSoup(response.content, 'html')
    print(soup.prettify())
    
    
    news_loop= soup.find_all('div',class_='news_loop') 
    news = {
        "headlines": [],
        "link": [],
        "body": [],
        "description": []
    }
    links = []
    for news in news_loop:
        link= news.find('a')['href']
        links.append(link)
    # links.pop(-2)
    # print(links)
    
    data = {
        'शीर्षक': [],
        'विवरण': [],
    }
    df = pd.DataFrame(data)
    
    
    for i in links:
        full_paragraph=''   
        response = requests.get(i)
        soup = BeautifulSoup(response.content, 'html.parser')
    
    
    
        title= soup.find('h1',class_="inside_head").text.strip()
        # title= soup.find('h1')
        
        
        print(title)
        bodyParagraph = soup.find_all('p')
        for paragraph in bodyParagraph:
            paragraph = paragraph.text.strip()
            
            words = paragraph.split()
            if any(re.search(r'[a-zA-Z]', word) for word in words):
                break
            if paragraph:
                if not paragraph.endswith("।"):
                    paragraph += "।"
                full_paragraph += paragraph + " "
              
        print(full_paragraph)
        print('\n\n')

        df.loc[len(df)] = [title, full_paragraph]
        

    
    today = date.today().strftime("%Y-%m-%d") 

    foldername='/var/home/ramrshrcg/Desktop/Python/scrapping/News_csv/dainiknepal'


    file_path = f"{foldername}/{today}_dainikNepal_news.csv"


    if os.path.exists(file_path):
        df.to_csv(file_path, mode='a', index=False, header=False)
    else:
        df.to_csv(file_path, index=False)



   
    
    
```
